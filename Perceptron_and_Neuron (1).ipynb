```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/placement-dataset.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
sns.scatterplot(x="cgpa", y="iq", data=df,hue=df['placement'])
```

```python
X= df[['cgpa','iq']]
y= df['placement']
```

```python
from sklearn.linear_model import Perceptron
model= Perceptron()
```

```python
model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values, y.values,clf=model, legend=2)
```
